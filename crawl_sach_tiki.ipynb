```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from time import sleep
import random
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome(service=Service("myenv/Lib/chromedriver.exe"))
driver.get('https://tiki.vn/sach-truyen-tieng-viet/c316?rating=4')
driver.maximize_window()
sleep(random.randint(3,10))
while True:
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 3);")
        sleep(2)
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//div[@data-view-id="category_infinity_view.more"]'))).click()
        sleep(2)
    except:
        break
sleep(5)
#books = driver.find_elements(books = driver.find_elements(By.XPATH, '//div[@class="styles__ProductItemContainerStyled-sc-bszvl7-0 elOGIo"]'))
books = WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH, '//div[@class="styles__ProductItemContainerStyled-sc-bszvl7-0 elOGIo"]')))
book_name, book_author, book_price, book_sold, book_link= [],[],[],[],[]
for book in books:
    try:
        name = book.find_element(By.XPATH, './/h3[@class="style__NameStyled-sc-139nb47-8 ibOlar"]').text
        book_name.append(name)
    except:
        book_name.append("")
    try:
        author = book.find_element(By.XPATH, './/div[@class="style__AboveProductNameStyled-sc-m30gte-0 hjPFIz above-product-name-info"]/span').text
        book_author.append(author)
    except:
        book_author.append("")
    try:
        price = book.find_element(By.XPATH, './/div[@class="price-discount__price"]').text
        book_price.append(price[:price.find("đ")])
    except:
        book_price.append("")
    try:
        sold = book.find_element(By.XPATH, './/span[@class="quantity has-border"]').text
        book_sold.append(sold[sold.find("n")+1:])
    except:
        book_sold.append("")
    try:
        link = book.find_element(By.XPATH, './/a[@class="style__ProductLink-sc-139nb47-2 cKoUly product-item"]').get_attribute("href")
        book_link.append(link)
    except:
        book_link.append("")
driver.quit()
df1 = pd.DataFrame(list(zip(book_name, book_author, book_price, book_sold, book_link)), columns=["BookName", "BookAuthor", "BookPrice", "BookSold", "BookLink"])
df1.to_csv("Book1.csv")
print(df1)
    

```

```python
def getlink(link):
    driver = webdriver.Chrome(service=Service("myenv/Lib/chromedriver.exe"))
    driver.get(link)
    sleep(random.randint(3,10))
    driver.maximize_window()
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight * 0.5);")
    sleep(2)
    try:
        adds = WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH, '//span[@class="show-more-content"]')))
        for add in adds:
            try:
                add.click()
                sleep(2)
            except:
                pass
    except:
        pass
    book_link, cus_name, thank_count, cus_status, cus_review, attributes = [],[],[],[],[],[]
    rev_card = driver.find_elements(By.XPATH, '//div[@class="style__StyledComment-sc-1y8vww-5 dpVjwc review-comment"]')
    for rev in rev_card:
        book_link.append(link)
        try:
            name = rev.find_element(By.XPATH, './/div[@class="review-comment__user-name"]').text
            cus_name.append(name)
        except:
            cus_name.append("")
        try:
            thank = rev.find_element(By.XPATH, './/div[@class="review-comment__user-info"]/span')[2].text
            thank_count.append(thank)
        except:
            thank_count.append("")
        try:
            status = rev.find_element(By.XPATH, './/div[@class="review-comment__title"]').text
            cus_status.append(status)
        except:
            cus_status.append("")
        try:
            review = rev.find_element(By.XPATH, './/div[@class="review-comment__content"/span').text
            cus_review.append(review)
        except:
            cus_review.append("")
        try:
            attribute = rev.find_element(By.XPATH, './/span[@class="rating-attribute__attributes"]').text
            attributes.append(attribute)
        except:
            attributes.append("")
        try:
            WebDriverWait(driver,10).until(By.XPATH, '//a[@class="btn next"]').click()
        except:
            pass
    df2 = pd.DataFrame(list(zip(book_link, cus_name, thank_count, cus_status, cus_review, attributes)), columns=["BookLink", "CustomerName", "ThankCount", "CustomerStatus", "CustomerReview", "Attributes"])
    return df2
```
