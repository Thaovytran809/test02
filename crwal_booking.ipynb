```python
def getreview(link):
    driver = webdriver.Chrome(service=Service("myenv/Lib/chromedriver.exe"))
    driver.get(link)
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 3);")
        try:
            WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[4]/div/div[5]/div[1]/div[1]/div[7]/div/div[3]/div/div[5]/button"))).click()
            break
        except:
            pass
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    sleep(random.randint(3,8))
    cus_name, cus_national, roomtype, cus_type, duration, posi_review, nega_review, cus_point, rev_title = [],[],[],[],[],[],[],[],[]
    while True:
        card_revs = driver.find_elements(By.XPATH, '//div[@data-testid="review-card"]')
        for card in card_revs:
            try:
                c_name = card.find_element(By.XPATH, './/div[@class="a3332d346a e6208ee469"]').text
                cus_name.append(c_name)
            except:
                cus_name.append("")
            try:
                c_national = card.find_element(By.XPATH, './/span[@class="afac1f68d9 a1ad95c055"]').text
                cus_national.append(c_national)
            except:
                cus_national.append("")
            try:
                c_type = card.find_element(By.XPATH, './/span[@data-testid="review-traveler-type"]').text
                cus_type.append(c_type)
            except:
                cus_type.append("")
            try:
                r_type = card.find_element(By.XPATH, './/span[@data-testid="review-room-name"]').text
                roomtype.append(r_type)
            except:
                roomtype.append("")
            try:
                day = card.find_element(By.XPATH, './/span[@data-testid="review-num-nights"]').text
                duration.append(day[:day.find("m")+1])
            except:
                duration.append("")
            try:
                posi_rev = card.find_element(By.XPATH, './/div[@data-testid="review-positive-text"]').text     
                posi_review.append(posi_rev)       
            except:
                posi_review.append("")
            try:
                nega_rev = card.find_element(By.XPATH, './/div[@data-testid="review-negative-text"]').text
                nega_review.append(nega_rev)
            except:
                nega_review.append("")
            try:
                c_point = card.find_element(By.XPATH, './/div[@class="ac4a7896c7"]').text[9:]
                cus_point.append(c_point)
            except:
                cus_point.append("")
            try:
                title = card.find_element(By.XPATH, './/div[@data-testid="review-title"]').text
                rev_title.append(title)
            except:
                rev_title.append("")
        try:
            next_button = driver.find_element(By.XPATH, '/html/body/div[4]/div/div[5]/div[1]/div[1]/div[1]/div[1]/div[4]/div/div[1]/div[2]/div/div[1]/div[2]/div/div/div[7]/div/div[2]/div[1]/div/div/div[3]/button')
            if next_button.is_enabled():
                next_button.click()
                sleep(3)
            else:
                break
        except (ElementNotInteractableException, NoSuchElementException):
            break
    driver.quit()
    df = pd.DataFrame(list(zip(cus_name,cus_national,cus_type,roomtype, duration, posi_review, nega_review, cus_point, rev_title)),columns=["CustomerName","CustomerNational","CustomerType","RoomType","Duration","Positive_Review","Negative_Review", "CustomerPoint", "Review_Title"])
    return df
    

```

```python
dfreview = pd.DataFrame()
for link in dfhotel["Links"]:
    df=getreview(link)
    if df.empty:
        continue
    h_name = dfhotel[dfhotel["Links"] == link]["HotelName"].values[0]
    df["HotelName"] = h_name
    dfreview = pd.concat([dfreview,df], ignore_index=True)
dfreview.to_csv("dfreview.csv", index=False)
print(dfreview)
```
