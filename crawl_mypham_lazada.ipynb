```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException, TimeoutException, ElementClickInterceptedException
from selenium.webdriver.common.by import By
from time import sleep
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import random
```

```python
driver = webdriver.Chrome(service=Service("myenv/Lib/chromedriver.exe"))
driver.get('https://www.lazada.vn/dung-cu-cham-soc-da-mat-2/?from=hp_categories&page=1&q=D%C6%B0%E1%BB%A1ng%20da&rating=4')
sleep(3)
pr_name, pr_price, pr_discount, pr_sold, pr_address, pr_rev_count, pr_link = [], [], [], [], [], [], []
while True:
    products = driver.find_elements(By.XPATH, '//div[@class="Bm3ON"]')
    for product in products:
        try:
            name = product.find_element(By.XPATH, './/div[@class="RfADt"]').text
            pr_name.append(name)
        except:
            pr_name.append("")
        try:
            link = product.find_element(By.TAG_NAME, 'a').get_attribute("href")
            pr_link.append(link)
        except:
            pr_link.append("")
        try:
            price = product.find_element(By.XPATH, './/span[@class="ooOxS"]').text
            pr_price.append(price[:price.find(" ")])
        except:
            pr_price.append("")
        try:
            discount = product.find_element(By.XPATH, './/span[@class="IcOsH"]').text
            pr_discount.append(discount[:discount.find(" ")])
        except:
            pr_discount.append("")
        try:
            sold = product.find_element(By.XPATH, './/span[@class="_1cEkb"]').text
            pr_sold.append(sold[:sold.find(" ")])
        except:
            pr_sold.append("")
        try:
            address = product.find_element(By.XPATH, './/span[@class="oa6ri "]').text
            pr_address.append(address)
        except:
            pr_address.append("")
        try:
            rev_count = product.find_element(By.XPATH, './/span[@class="qzqFw"]').text
            pr_rev_count.append(rev_count[rev_count.find("(")+1:rev_count.find(")")])
        except:
            pr_rev_count.append("")
    try:
        next_page = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//li[@class="ant-pagination-next"]/button')))
        next_page.click()
        sleep(3)
        #WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[3]/div/div[2]/div[1]/div/div[1]/div[4]/div/ul/li[9]/button'))).click()
    except (NoSuchElementException, ElementNotInteractableException, TimeoutException):
        break
driver.quit()
dfproduct = pd.DataFrame(list(zip(pr_name, pr_price, pr_discount, pr_sold, pr_address, pr_rev_count, pr_link)), columns=["ProductName", "ProductPrice", "ProductDiscount", "ProductSold", "ProductAddress", "ProductReviewCount", "Productlink"])
print(dfproduct)
dfproduct.to_csv("dfproduct.csv", index=False)
        
```

```python
dfproduct = pd.read_csv("dfproduct.csv")
```

```python
def getreview(link):
    random.randint(3,10)
    driver = webdriver.Chrome(service=Service("myenv/Lib/chromedriver.exe"))
    driver.get(link)
    driver.maximize_window()
    sleep(random.randint(3,5))
    cus_name, pr_brand, pr_catagory, pr_review = [],[],[],[]
    while True:
        try:
            robot = WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, '//img[@data-spm-anchor-id="a2o4n.pdp_revamp.0.i0.589e6bebO4z6NA]')))
            robot.click()
        except:
            pass
        reviews = driver.find_elements(By.XPATH, '//div[@class="item"]')
        for rev in reviews:
            try:
                name = rev.find_element(By.XPATH, './/div[@class="middle"]').text
                cus_name.append(name[:name.find("Chứng")])
            except:
                cus_name.append("")
            try:
                catagory = rev.find_element(By.XPATH, './/div[@class="skuInfo"]').text
                pr_catagory.append(catagory[catagory.find(":")+1:])
            except:
                pr_catagory.append("")
            try:
                review = rev.find_element(By.XPATH, './/div[@class="content"]').text
                pr_review.append(review)
            except:
                pr_review.append("")
            try:
                brand = driver.find_element(By.XPATH, './/a[@class="pdp-link pdp-link_size_s pdp-link_theme_blue pdp-product-brand__brand-link"]').text
                pr_brand.append(brand)
            except:
                pr_brand.append("")
        try:
            next_page = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//button[@class="next-btn next-btn-normal next-btn-medium next-pagination-item next"]')))
            next_page.click()
            sleep(3)
        except (NoSuchElementException, ElementNotInteractableException, TimeoutException, ElementClickInterceptedException) :
            break
    driver.quit()
    df = pd.DataFrame(list(zip(cus_name, pr_brand, pr_catagory, pr_review)), columns=["CustomerName", "ProductBrand", "ProductCatagory", "ProductReview"])
    return df
        


```

```python
dfdanhgia = pd.DataFrame()
d=0
for link in dfproduct["Productlink"]:
    d=d+1
    df = getreview(link)
    if df.empty:
        continue
    p_name = dfproduct[dfproduct["Productlink"]==link]["ProductName"].values[0]
    df["ProductName"] = p_name
    df["Link"] = link
    dfdanhgia = pd.concat([dfdanhgia,df], ignore_index=True)
    if d==5:
        break
print(dfdanhgia)
dfdanhgia.to_csv("dfdanhgia.csv")

```
